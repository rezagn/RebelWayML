```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("D:\Courses\RebelWay\Introduction to Machine Learning\Assignments\Week04\\breast_cancer.csv")
```

```python
df.drop('Unnamed: 32', axis = 1, inplace = True)
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
sns.countplot(x = 'diagnosis', data = df, palette = 'RdBu_r', hue = 'diagnosis')
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('diagnosis', axis = 1), df['diagnosis'], test_size = 0.30, random_state = 101)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
print(predictions)
```

```python
print(classification_report(y_test, predictions))
```
