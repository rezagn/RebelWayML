```python
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = "viridis")
```

```python
# Filling and Removing NAN values
mean_age = df["Age"].mean()
df.fillna({"Age":mean_age}, inplace = True)
df.dropna(axis = 1, inplace = True)
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = "viridis")
```

```python
sns.set_style("whitegrid")
sns.countplot(x="Survived", data = df, palette = "RdBu_r", hue = "Survived", legend = False)
```

```python
sns.countplot(x = "Survived", data = df, hue = "Sex", palette = "RdBu_r", legend = False)
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(x = "Sex", y = "Age", data = df, palette = "summer", hue = "Sex", legend = False)
```

```python
sns.displot(df["Age"], kde = True, color = "darkblue", bins = 50)
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left = True, bottom = True)
sns.scatterplot(x = "Age", y = "Fare", hue = "Sex", linewidth = 2, data = df, ax = ax)
```
