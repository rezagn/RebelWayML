```python
# Using the pyplot API
plt.plot(df['Age'], df['Fare'], "g.")
plt.title("Simple Age/Fare")
plt.show()
```

```python

# Axes API
fig, ax = plt.subplots(figsize = (5.5, 3), layout = 'constrained')
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_title("Axes Age/Fare")
#ax.plot(df['Age'], df['Fare'], "g.")
ax.scatter(df['Age'], df['Fare'], c = 'c', s = 1)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = 'polar')
c = ax.scatter(x = df['Fare'], y = df['Age'], alpha = 0.25, c = 'g')
```

```python
#fig = plt.figure()
#ax = fig.add_subplot(projection = 'rectilinear')
survived = df['Survived'].value_counts()
print(survived)
gender = df['Sex'].value_counts()
print(gender)
male_survived = df[(df['Sex'] == "male") & (df['Survived'] == 1)].shape[0]
male_died = df[(df['Sex'] == "male") & (df['Survived'] == 0)].shape[0]
female_survived = df[(df['Sex'] == "female") & (df['Survived'] == 1)].shape[0]
female_died = df[(df['Sex'] == "female") & (df['Survived'] == 0)].shape[0]
rates = pd.Series([male_survived, male_died, female_survived, female_died])
print(rates)

#c = ax.scatter(df['Sex'], df['Survived'])
```

```python

plt.style.use('_mpl-gallery-nogrid')

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(rates)))

fig, ax = plt.subplots()
ax.pie(rates, colors = colors, radius = 4.5, center = (5, 5), wedgeprops = {"linewidth": 1, "edgecolor": "white"}, frame = True)
ax.set(xlim = (0, 10), xticks = np.arange(1, 10),
       ylim = (0, 10), yticks = np.arange(1, 10))

plt.show()
```
